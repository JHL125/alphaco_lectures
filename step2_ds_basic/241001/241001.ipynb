## AAPL 주식, NVDA주식 두개 불러와서 동시에 라인차트 그려보기

```python
import yfinance as yf
apple = yf.download("AAPL", start="2020-01-01", end="2024-09-30")
nvidia = yf.download("NVDA", start="2020-01-01", end="2024-09-30")

apple.head(1)
```

```python
nvidia.head(1)
```

```python
fig, ax = plt.subplots(figsize=(5, 3)) # 차트 사이즈

ax.plot(apple['Open'], label="Apple")
ax.plot(nvidia['Open'], label="NVDA")
ax.set_title("Stocks") #제목 붙이기
ax.legend() #범례추가 (옆에다가)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(10,4)) 

ax[0].plot(apple['Open'], label="Apple") #0,1 로 두개 만emfrl
ax[1].plot(nvidia['Open'], label="NVDA")

ax[0].set_title("apple stocks") #제목 붙이기
ax[1].set_title("nvda stocks")

ax[0].legend()
ax[1].legend()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,4)) 

ax[0, 0].plot(apple['Open'], label="Apple") #0,1 로 두개 만들기
ax[0, 1].plot(nvidia['Open'], label="NVDA")
ax[1, 0].plot(intel['Open'], label="intel")

ax[0, 0].set_title("apple stocks") #제목 붙이기
ax[0, 1].set_title("nvda stocks")
ax[1,0].set_title("intel stocks")

ax[0, 0].legend()
ax[0, 1].legend()
ax[1, 0].legend()

fig.tight_layout() # 간격벌리기
plt.savefig("myStocks.png") # 그림으로 내보내기, 저장해서 출력!  
plt.show()
```
